```
from __future__ import absolute_import
from __future__ import print_function
import numpy 
import numpy as np
import os
import keras
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras import Sequential
from keras.preprocessing import sequence
from keras.models import Sequential, Model
from keras.layers import Dense, Dropout, Bidirectional,Input,Activation,Flatten
from keras.datasets import imdb
from keras import backend as K
from keras.layers import Conv1D, GlobalAveragePooling1D, MaxPooling1D,GlobalMaxPooling1D
import typing
from keras.utils.np_utils import to_categorical
from keras import Model
from keras import layers
from keras import Input
num_classes = 5
epochs = 2
```

```
def create_base_network(input_shape):  
 
   input = Input(shape = input_shape)
   x = Conv1D(64,1, activation='relu' )(input)
   #x = Dropout(0.5)(x)
   #x = MaxPooling1D(pool_size=1)(x)
   #x = Conv1D(64,1, activation='relu')(x) 
   #x = Dropout(0.5)(x)
   #x = MaxPooling1D(pool_size=1)(x)
   #x = Conv1D(64,1, activation='relu')(x) 
   return Model(input,x)

dataset = numpy.loadtxt("trx1.csv", delimiter=",",dtype= 'int')
dataset2 = numpy.loadtxt("trx2.csv", delimiter=",",dtype= 'int')
datasety = numpy.loadtxt("datay.csv", delimiter=",",dtype= 'int')
datasete = numpy.loadtxt("tex1.csv", delimiter=",",dtype= 'int')
datasete2 = numpy.loadtxt("tex2.csv", delimiter=",",dtype= 'int')
datasetye = numpy.loadtxt("tey.csv", delimiter=",",dtype= 'int')

A = np.resize(dataset , (224,1,900))
print("x_train",A.shape)
print("x_train[1:]",A.shape[1:])
B = np.resize(dataset2 , (224,1,900))
print(B.shape)
C = np.resize(datasety , (224,1))
print(C.shape)

pair = np.vstack((A,B))   # vertical stack

pair = np.resize(pair , (224,2,1,900))
print("pair",pair.shape)
print("pair[:,0]",pair[:,0].shape)
print("pair[:,1]",pair[:,1].shape)

a = np.resize(datasete , (57,1,900))
print(a.shape)
b = np.resize(datasete2 , (57,1,900))
print(b.shape)
c = np.resize(datasetye , (57,1))
print(c.shape)

paire = np.vstack((a,b))   # vertical stack

paire = np.resize(paire , (57,2,1,900))

print("paire",paire.shape)
print("paire1",paire[:,0].shape)
print("paire2",paire[:,1].shape)

y1= keras.utils.to_categorical(C,num_classes=5)
y2= keras.utils.to_categorical(c,num_classes=5)
y1c = np.resize(y1 , (224,5,1))
y2c = np.resize(y2 , (57,5,1))
print("y1",y1.shape)
kernel_size = 3
hidden_dims = 64
epochs = 20

input_shape = A.shape[1:]
base_network = create_base_network(input_shape)

input_a = Input(shape=input_shape)
input_b = Input(shape=input_shape)

processed_a = base_network(input_a)
processed_b = base_network(input_b)

answer_size = 5

concatenate = layers.concatenate([processed_a, processed_b],axis = -1)
y3 = Flatten()(concatenate)
lasty = layers.Dense(answer_size,
                      activation = 'softmax')(y3)
model = Model([input_a, input_b], lasty) 
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
model.fit([pair[:, 0], pair[:, 1]], y1,
          batch_size=1,
          epochs=epochs,
          validation_data=([paire[:, 0], paire[:, 1]], y2))
```
