Exercise 3 Synthesize a compound signal by creating SinSignal and CosSignal objects and adding them up. Evaluate the signal to get a Wave, and listen to it. Compute its Spectrum and plot it. What happens if you add frequency components that are not multiples of the fundamental?

```python
from __future__ import print_function, division

%matplotlib inline
#%pip install thinkx

import thinkdsp
import thinkplot

import warnings
warnings.filterwarnings('ignore')

from IPython.html.widgets import interact, fixed
from IPython.display import display

wave = thinkdsp.read_wave('36179__erh__thunder-pad.wav')
wave.normalize()
wave.make_audio()

wave.plot()

import matplotlib.pyplot as pl
pl.grid()
```

```python

signal = (thinkdsp.SinSignal(freq=400, amp=1.0) +
          thinkdsp.SinSignal(freq=600, amp=0.5) +
          thinkdsp.SinSignal(freq=800, amp=0.25))
signal.plot()
```

```python

wave2 = signal.make_wave(duration=1)
wave2.apodize()
```

```python
wave2.make_audio()
```

```python

spectrum = wave2.make_spectrum()
spectrum.plot(high=2000)
```

```python

signal += thinkdsp.SinSignal(freq=450)
signal.make_wave().make_audio()
```
